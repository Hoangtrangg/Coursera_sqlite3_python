```python
import os
import sqlite3 as sql


# generating a list of values to insert to database
values = [] 

for root, director, file in os.walk('.'):
    for f in file:
        file_path = os.path.join(root,f)
        size = os.path.getsize(file_path)
        values.append([file_path,size])

print(values)
```

```python
# create connection and database
connection = sql.connect('filesize.db')

create_table_query = 'create table filesize (id integer primary key, path text, size integer)'

cursor = connection.cursor()
cursor.execute(create_table_query)
connection.commit()
```

```python
# insert data to table
insert_query = 'insert into filesize(path, size) values(?,?)'

for item in values:
    cursor.execute(insert_query,item)

connection.commit()
```

```python
# select item in the table
for i in cursor.execute('select * from filesize'):
    print(i)
```

```python
# select 20 biggest file
query = 'select path, size from filesize order by size desc limit(20)'
for i in cursor.execute(query):
    print(i)
```

```python
#select files having size between 3000000 and 5000000 bytes
query = 'select path, size from filesize where size >3000000 and size < 5000000 order by size asc'
for i in cursor.execute(query):
    print(i)
```
